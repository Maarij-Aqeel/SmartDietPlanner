```python
def create_conversation(row):
  user_query = random.choice(user_queries).format(
        Diet_type=row['Diet_type'],
        Cuisine_type=row['Cuisine_type'],
        Protein=row['Protein(g)'],
        Carbs=row['Carbs(g)'],
  )

  if "high in protein" in user_query:
      assistant_response = (
          f"I'd recommend '{row['Recipe_name']}', a {row['Diet_type']} recipe with "
          f"{row['Protein(g)']}g of protein, perfect for meeting your protein needs. Would you like the recipe instructions?"
      )
  elif "low-carb" in user_query:
      assistant_response = (
          f"'{row['Recipe_name']}' is a great low-carb option from {row['Cuisine_type']} cuisine, "
          f"with only {row['Carbs(g)']}g of carbs and {row['Protein(g)']}g of protein. It's a healthy choice."
      )
  elif "most protein" in user_query:
      assistant_response = (
          f"'{row['Recipe_name']}' from the {row['Diet_type']} diet has the highest protein content, "
          f"offering {row['Protein(g)']}g of protein. It's the best option if you want to build muscle."
      )
  elif "less than" in user_query and "carbs" in user_query:
      assistant_response = (
          f"You could try '{row['Recipe_name']}', a {row['Diet_type']} recipe with only {row['Carbs(g)']}g of carbs. It's great for a quick light meal"
      )
  elif "more than" in user_query and "protein" in user_query:
      assistant_response = (
          f"'{row['Recipe_name']}' is a {row['Cuisine_type']} dish that packs {row['Protein(g)']}g of protein, "
          f"making it a protein-rich option."
      )
  elif "least amount of fat" in user_query:
      assistant_response = (
          f"'{row['Recipe_name']}' has the least amount of fat among the {row['Diet_type']} recipes, "
          f"containing only {row['Fat(g)']}g of fat."
      )
  elif "low in carbs and high in protein" in user_query:
      assistant_response = (
          f"'{row['Recipe_name']}' fits the bill perfectly: it has {row['Carbs(g)']}g of carbs and "
          f"{row['Protein(g)']}g of protein."
      )
  elif "lowest carbs" in user_query:
      assistant_response = (
          f"In {row['Cuisine_type']} cuisine, '{row['Recipe_name']}' has the lowest carb count, "
          f"with just {row['Carbs(g)']}g of carbs."
      )
  elif "rich in fat" in user_query:
      assistant_response = (
          f"If you're looking for fat-rich but low-carb recipes, '{row['Recipe_name']}' is ideal, with "
          f"{row['Fat(g)']}g of fat and {row['Carbs(g)']}g of carbs. It's a great option."
      )
  else:
      assistant_response = (
          f"'{row['Recipe_name']}' is a {row['Diet_type']} recipe with "
          f"{row['Protein(g)']}g of protein, {row['Carbs(g)']}g of carbs, and {row['Fat(g)']}g of fat."
      )

  return {'query':user_query, 'response':assistant_response }
```

# A Smarter Approach

Modifying the code to take into account the current row context, instead of randomly picking rows

```python
import pandas as pd
import random

df = pd.read_csv("All_Diets.csv", usecols=["Diet_type", "Recipe_name", "Cuisine_type", "Protein(g)", "Carbs(g)", "Fat(g)"])
```

```python
df.columns = df.columns.str.replace('(', '_').str.replace(')', '')

# Define thresholds for nutritional criteria
PROTEIN_THRESHOLD = 20  # grams
CARB_THRESHOLD = 15     # grams
FAT_THRESHOLD = 10      # grams

# Add binary columns for high/low nutritional values
df['high_protein'] = df['Protein_g'] > PROTEIN_THRESHOLD
df['low_carb'] = df['Carbs_g'] < CARB_THRESHOLD
df['low_fat'] = df['Fat_g'] < FAT_THRESHOLD
```

```python
user_queries = [
    "Can you recommend a {Diet_type} recipe that's high in protein?",
    "What's a low-carb recipe from {Cuisine_type} cuisine?",
    "Which recipe in the {Diet_type} diet contains the most protein?",
    "Can you suggest a {Diet_type} recipe with less than {Carbs_g}g of carbs?",
    "I'm looking for a {Cuisine_type} recipe that has more than {Protein_g}g of protein.",
    "What recipe in the {Diet_type} diet has the least amount of fat?",
    "What recipe from {Cuisine_type} cuisine has the lowest carbs?",
    "I need a recipe that's rich in fat but low in carbs for the {Diet_type} diet.",
    "Show me a low-carb recipe from the {Diet_type} diet.",
    "Give me a recipe that's low in carbs and high in protein."
    "Give me a {Cuisine_type} recipe with the lowest carbs."
]
```

```python
import re

def get_relevant_rows(user_query, df):
    relevant_rows = df

    if any(phrase in user_query for phrase in ["high in protein", "high-protein", "more protein"]):
        relevant_rows = df[df['high_protein']]

    elif any(phrase in user_query for phrase in ["low-carb", "low carbs"]):
        relevant_rows = df[df['low_carb']]

    elif "most protein" in user_query:
        relevant_rows = df[df['Protein_g'] == df['Protein_g'].max()]

    elif "less than" in user_query and "carbs" in user_query:
        carb_threshold = extract_numeric_value(user_query)
        if carb_threshold is not None:
            relevant_rows = df[df['Carbs_g'] < carb_threshold]

    elif "more than" in user_query and "protein" in user_query:
        protein_threshold = extract_numeric_value(user_query)
        if protein_threshold is not None:
            relevant_rows = df[df['Protein_g'] > protein_threshold]

    elif "least amount of fat" in user_query:
        relevant_rows = df[df['Fat_g'] == df['Fat_g'].min()]

    elif "low in carbs and high in protein" in user_query:
        relevant_rows = df[df['low_carb'] & df['high_protein']]

    elif "rich in fat" in user_query:
        relevant_rows = df[~df['low_fat']]

    else:
        relevant_rows = None

    return relevant_rows


def extract_numeric_value(query):
    """Extract the first numeric value from a user query."""
    match = re.search(r"(\d+\.?\d*)", query)
    if match:
        return float(match.group(1))
    return None
```

```python
def create_conversations(row):
  num_queries = random.randint(2, 3)
  conversations = []
  for _ in range(num_queries):
    user_query = random.choice(user_queries).format(
        Diet_type=row['Diet_type'],
        Cuisine_type=row['Cuisine_type'],
        Protein_g=row['Protein_g'],
        Carbs_g=row['Carbs_g'],
        Fat_g=row['Fat_g']
    )

    relevant_rows = get_relevant_rows(user_query, df)
    if relevant_rows is not None and not relevant_rows.empty:
      selected_row = relevant_rows.sample().iloc[0]
    else:
      selected_row = row

    if "high in protein" in user_query:
        assistant_response = (
            f"I'd recommend '{selected_row['Recipe_name']}', a {selected_row['Diet_type']} recipe with "
            f"{selected_row['Protein_g']}g of protein, perfect for meeting your protein needs."
        )
    elif "low-carb" in user_query:
        assistant_response = (
            f"'{selected_row['Recipe_name']}' is a great low-carb option from {selected_row['Cuisine_type']} cuisine, "
            f"with only {selected_row['Carbs_g']}g of carbs and {selected_row['Protein_g']}g of protein. Consistency is key, and this will help!"
        )
    elif "most protein" in user_query:
        assistant_response = (
            f"Lets fuel your body with this awesome choice: '{selected_row['Recipe_name']}' from the {selected_row['Diet_type']} diet has the highest protein content, "
            f"offering {selected_row['Protein_g']}g of protein."
        )
    elif "less than" in user_query and "carbs" in user_query:
        assistant_response = (
            f"You could try '{selected_row['Recipe_name']}', a {selected_row['Diet_type']} recipe with only {selected_row['Carbs_g']}g of carbs."
        )
    elif "more than" in user_query and "protein" in user_query:
        assistant_response = (
            f"'{selected_row['Recipe_name']}' is a {selected_row['Cuisine_type']} dish that packs {selected_row['Protein_g']}g of protein, "
            f"making it a protein-rich option."
        )
    elif "least amount of fat" in user_query:
        assistant_response = (
            f"'{selected_row['Recipe_name']}' has the least amount of fat among the {selected_row['Diet_type']} recipes, "
            f"containing only {selected_row['Fat_g']}g of fat. You're going to enjoy this one, trust me!"
        )
    elif "low in carbs and high in protein" in user_query:
        assistant_response = (
            f"'{selected_row['Recipe_name']}' fits the bill perfectly: it has {selected_row['Carbs_g']}g of carbs and "
            f"{selected_row['Protein_g']}g of protein. This will keep you on track and feeling great"
        )
    elif "rich in fat" in user_query:
        assistant_response = (
            f"If you're looking for fat-rich but low-carb recipes, '{selected_row['Recipe_name']}' is ideal, with "
            f"{selected_row['Fat_g']}g of fat and {selected_row['Carbs_g']}g of carbs. Its packed with the nutrients you need!"
        )
    else:
        assistant_response = (
            f"Ive got a fantastic suggestion for you: '{row['Recipe_name']}' is a {row['Diet_type']} recipe with "
            f"{row['Protein_g']}g of protein, {row['Carbs_g']}g of carbs, and {row['Fat_g']}g of fat. Keep it upyoure doing great!"
        )

    conversations.append({
            'query': user_query,
            'response': assistant_response
        })

  return conversations

```

```python
all_conversations = []
for _, row in df.iterrows():
    all_conversations.extend(create_conversations(row))

conversation_df = pd.DataFrame(all_conversations)
```

```python
import pandas as pd

df = pd.read_csv('recipes_filtered.csv')

relevant_columns = ['recipe_name', 'prep_time', 'cook_time', 'ingredients', 'servings', 'directions', 'nutrition']
df_relevant = df[relevant_columns]
```

```python
import pandas as pd
import re

# list of unnecessary prefixes
prefixes = [
    "Sarah's", "Mom's", "Easy", "Amazing", "Awesome", "Best", "Quick", "Simple",
    "Grandma's", "The Best", "World's Best", "Ma Lipo's"]

# Create a regex pattern for prefixes that match at the start of the string
pattern = r"^(?:" + "|".join(re.escape(prefix) for prefix in prefixes) + r")\s+"

def clean_recipe_name(name):
    if isinstance(name, str):
        return re.sub(pattern, "", name, flags=re.IGNORECASE).strip()
    return name

df_relevant['recipe_name'] = df_relevant['recipe_name'].apply(clean_recipe_name)

print(df_relevant[['recipe_name']].head())
```

```python
user_queries = [
    "What are the steps to prepare {recipe_name}?",
    "How long does it take to prepare {recipe_name}?",
    "What ingredients do I need to make {recipe_name}?",
    "How many servings does {recipe_name} yield?",
    "Can you give me the full recipe for {recipe_name}?",
    "What is the nutrition information for {recipe_name}?",
    "How do I make {recipe_name}?",
    "Can you break down the steps to cook {recipe_name}?",
    "What is the prep and cook time for {recipe_name}?",
]
```

```python
import random

def create_response(row, query):
    if any(f in query for f in ["steps", "how to make", "instructions", "directions", "make"]):
        return "To prepare {}, follow these steps: \n{}".format(row['recipe_name'], '\n'.join(row['directions'].split('.')))

    elif any(f in query.lower() for f in ["prep time", "cook time", "total time", "how long"]):
        return "{} usually takes {} minutes to prepare and {} minutes to cook.".format(row['recipe_name'], row['prep_time'], row['cook_time'])

    elif "ingredients" in query:
        return "You will need the following ingredients to make {}: {}\nNote that you can also substitute some ingredients for others of your preference.".format(row['recipe_name'], row['ingredients'])

    elif any(f in query for f in ["servings", "yield"]):
        serving_type = "individual" if row['servings'] < 2 else "group"
        return "{} yields {} servings. This is great for {} serving meal.".format(row['recipe_name'], row['servings'], serving_type)

    elif any(f in query for f in ["nutrition", "nutrients", "calories", "macros"]):
        nutrition_info = ",\n".join(row['nutrition'].split(','))
        return "Here is the nutrition information for {}: {}".format(row['recipe_name'], nutrition_info)

    elif "recipe" in query:
        nutrition_info = ",\n".join(row['nutrition'].split(','))
        return "Here is the full recipe for {}:\nIngredients: {}\nServings: {}\nPrep time: {}\nCook time: {}\nDirections: {}\nNutrition: {}".format(
            row['recipe_name'],
            ''.join(row['ingredients']),
            row['servings'],
            row['prep_time'],
            row['cook_time'],
            ''.join(row['directions']),
            nutrition_info
        )

    return "I don't have a specific response for this query."
```

```python
def create_conversation(row):
    num_queries = random.randint(2, 5)
    conversations = []
    for _ in range(num_queries):
      query = random.choice(user_queries).format(recipe_name=row['recipe_name'])
      response = create_response(row, query)
      conversations.append({
            'query': query,
            'response': response
        })

    return conversations
```
